# **Get an initial understanding of data**

Import necessery libraries.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### **Add dataset and get an initial understanding**

As a first step, the below code `pd.read_csv()` reads a CSV file into a pandas DataFrame. and get a some idea about the data.

```python
datafile = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/WA_Fn-UseC_-Telco-Customer-Churn.csv')
datafile
```

show the info about data using `datafile.info()`.

```python
datafile.info()
```

### **Calculate basic summarization statistics**

See the summery of all the numerical dataset in the datafile using `datafile.describe()`.

```python
numerical_summary = datafile.describe()
numerical_summary
```

See the summery of all the datatype of object in the dataset using `datafile.describe(include=['object')`.

```python
categorical_summary = datafile.describe(include=['object'])
categorical_summary
```

### **Visualize the data**

```python

datafile.hist(figsize=(15, 10))
plt.show()


for column in datafile.drop(columns='customerID').select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(datafile[column])
    plt.xticks(rotation=90)
    plt.show()

correlation_matrix = datafile.select_dtypes(include=['int64', 'float64']).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```
